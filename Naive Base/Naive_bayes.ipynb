```python
data_x = data[:, 1:]
data_y = data[:, 0]
```

```python
split = 0.8

train_x = data_x[:int(split*data_x.shape[0]), :]
train_y = data_y[:int(split*data_y.shape[0])]


test_x = data_x[int(split*data_x.shape[0]):, :]
test_y = data_y[int(split*data_y.shape[0]):]
```

```python
print(train_x.shape,train_y.shape, test_x.shape, test_y.shape )
```

```python
def prior_prob(y_train, label):
    prob = np.sum(y_train == label)/y_train.shape[0]
    return prob
```

```python
def likelihood_prob(y_train, x_train, col, value, label):
    x_train = x_train[y_train == label]
    numerator = np.sum(x_train[:, col] == value)
    denominator = np.sum(y_train == label)
    return numerator/denominator
        
```

```python
def posterior_prob(x_train, y_train, x_test):
    num_classes = np.unique(y_train)
    posterior = []
    
    
    for label in num_classes:
        prior = prior_prob(y_train, label)
        likelihood = 1
        for features in range(x_train.shape[1]):
            likelihood *= likelihood_prob(y_train, x_train, features, x_test[features], label)
            
        posterior.append(likelihood * prior)
    pred = np.argmax(posterior)
    return pred
```

```python
corr = 0

for i in range(1625):
    #print(i)
    print(posterior_prob(train_x, train_y, test_x[i]), test_y[i])
    if posterior_prob(train_x, train_y, test_x[i]) == test_y[i]:
        corr = corr + 1

print(corr/test_y.shape[0])
```
