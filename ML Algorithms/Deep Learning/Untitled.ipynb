```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
```

```python
from keras.datasets import boston_housing
```

```python
(train_x, train_y), (test_x, test_y) = boston_housing.load_data()
```

```python
print(train_x.shape, train_y.shape, test_x.shape, test_y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
train_x = min_max_scaler.fit_transform(train_x)
test_x = min_max_scaler.fit_transform(test_x)
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.models import Sequential
```

```python
model = Sequential(name = 'Bostan_ANN')
model.add(Dense(500, input_dim=train_x.shape[1], activation='relu', name='Input_Layer'))
model.add(Dense(400, activation='relu', name='2_Layer'))
model.add(Dense(300, activation='relu', name='3_Layer'))
model.add(Dense(200, activation='relu', name='4_Layer'))
model.add(Dense(100, activation='relu', name='5_Layer'))
model.add(Dense(50, activation='relu', name='6_Layer'))
model.add(Dense(1, activation='linear', name='Output_Layer'))
```

```python
model.compile(optimizer = 'Adam', loss = 'mse', metrics = [keras.metrics.RootMeanSquaredError()])
```

```python
model.summary()
```

```python
hist = model.fit(train_x, train_y, epochs=1000, batch_size= 500, validation_data=(test_x, test_y))
```

```python
hist.history
```

```python
score, acc = model.evaluate(test_x, test_y)
```
