```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
(trainX, trainY), (testX, testY) = keras.datasets.mnist.load_data()
```

```python
print('The shape of the training inputs:', trainX.shape)
print('The shape of the training labels:',trainY.shape)
print('The shape of the testing inputs:',testX.shape)
print('The shape of the testing labels:',testY.shape)
```

```python
input_dimnetion = (28,28)
```

```python
# plotting the first 16 images in the train set of MNIST
 
fig, axs = plt.subplots(4, 4)
cnt = 0
for i in range(4):
     for j in range(4):
         axs[i, j].imshow(trainX[cnt])
         cnt += 1
```

```python
# Normolize the dataset

trainX = tf.keras.utils.normalize(trainX, axis=1)
testX = tf.keras.utils.normalize(testX, axis=1)
```

```python
# BUILDING THE NEURAL NETWORK

# INPUT LAYER
model = Sequential(name = 'Sequential MNIST Digit Aataset')
model.add(tf.keras.layers.Flatten(input_shape = (28,28))

model.add(Dense(400, activation='relu', name='2_Layer'))
model.add(Dense(300, activation='relu', name='3_Layer'))
model.add(Dense(200, activation='relu', name='4_Layer'))
model.add(Dense(100, activation='relu', name='5_Layer'))

# OUTPUT LAYER

model.add(Dense(10, activation='softmax', name='Output_Layer'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss=tf.keras.losses.CategoricalCrossentropy(), metrics=['accuracy', 'Recall', 'Precision'])
```

```python
hist = model.fit(trainX, trainY, batch_size=100, epochs=10)
```
