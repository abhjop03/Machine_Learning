```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
data.keys()
```

```python
df_data = pd.DataFrame(data['data'], columns=data['feature_names'])
```

```python
df_label = pd.DataFrame(data['target'], columns = ['label'])
```

```python
df_label
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df_data, df_label, shuffle=True, test_size=0.2 )
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel = 'poly', degree =2)
svm_classifier.fit(X_train, Y_train)
```

```python
pred = svm_classifier.predict(X_test)
```

```python
pred
```

```python
print(classification_report(Y_test, pred))
```

```python
from sklearn.model_selection import GridSearchCV

params= {
    'kernel' : ['rbf', 'linear', 'sigmoid'],
    'C' : [1,0.5, 0.1, 0.05, 0.01, 0.005, 0.001],
    'gamma' : [1, 0.1, 0.01, 0.001]
}
```

```python
grid = GridSearchCV(svm.SVC(), param_grid= params, return_train_score= True)
grid.fit(X_train, Y_train)
```

```python
grid.cv_results_['params']
```

```python
grid.best_estimator_
grid.best_params_
```

```python
from sklearn import svm

svm_classifier = svm.SVC(kernel = 'linear', C = 1, gamma= 1)
svm_classifier.fit(X_train, Y_train)
```

```python
pred = svm_classifier.predict(X_test)
print(classification_report(Y_test, pred))
```
