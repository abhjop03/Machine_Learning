```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
data, label = make_blobs(n_samples=1000, centers=5, n_features=2, random_state=500)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
data.shape, label.shape
```

```python
plt.scatter(data[:,0], data[:,1], c=label)
```

```python
kmeans.fit(data)
```

```python
centroids = kmeans.cluster_centers_  # it tells that the centroid points of clusters.
```

```python
labels = kmeans.labels_
```

```python
print("Centroids:")
print(centroids)
print("\nCluster Labels:")
print(labels)
```

```python
def plot_scatter():
    plt.scatter(data[:,0], data[:,1], c=label)
    plt.scatter(centroids[:,0], centroids[:,1], marker='*', c='black')
```

```python
plot_scatter()
```

```python
# Inplementing the elbow concept to find the optimum no of clusters.
elbow = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data)
    elbow.append(kmeans.inertia_)
```

```python
plt.plot(elbow)
```
