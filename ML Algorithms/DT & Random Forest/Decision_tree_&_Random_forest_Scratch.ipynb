```python
def entropy(cols):
    favour = np.unique(cols, return_counts= True)
    total = cols.shape[0]
    ent = 0
    for i in favour[1]:
        prob = i/total
        ent = -1*(prob * np.log2(prob))
    return ent
        
```

```python
def devide_data(x_data, feat, threshold):
    x_right = pd.DataFrame([],columns= x_data.columns)
    x_left = pd.DataFrame([],columns= x_data.columns)
    for i in range(x_data.shape[0]):
        value = x_data[feat].loc[i]
        
        if value >= threshold:
            x_right = pd.concat([x_right, x_data.iloc[i : i+1]])
        else:
            x_left = pd.concat([x_left, x_data.iloc[i : i+1]])
    return x_left, x_right
```

```python
def info_gain(x_data, feat, threshold):
    left, right = devide_data(x_data, feat, threshold)

    rho_l = left.shape[0] / x_data.shape[0]
    rho_r = right.shape[0] / x_data.shape[0]
    i_gain = entropy(x_data.Survived) - (rho_l * entropy(left.Survived) + rho_r * entropy(right.Survived))
    return i_gain
```

```python
for i in x_data.columns:
    print(i, info_gain(data, i, data[i].mean()))
```

```python
class Decision_tree:

    def __init__(self, max_depth = 10, cur_depth = 0):
        self.left = None
        self.right = None
        self.feat = None
        self.value = None
        self.depth = cur_depth
        self.max_depth = max_depth
        self.target = None

    def train_dt(self, x_data):
        temp = []
        f = ['Pclass','Sex','Age','SibSp','Parch','Fare', 'Embarked']
        i_gain = []

        for i in f:
            i_gain.append(info_gain(x_data, i, x_data[i].mean()))
        self.feat = f[np.argmax(i_gain)]
        self.value = x_data[self.feat].mean()
        left,right = devide_data(x_data, self.feat, self.value)
        left = left.reset_index(drop = True)
        right = right.reset_index(drop = True)
        temp.append(f'Making Node of {self.feat}')
        print(f'Making Node of {self.feat}')
        if left.shape[0] == 0 or right.shape[0] == 0:
            if x_data.Survived.mean(0) >= 0.5:
                self.target = 'Survived'
            else:
                self.target = 'Not Survived'
            return
        if self.depth >= self.max_depth:
            if x_data.Survived.mean(0) >= 0.5:
                self.target = 'Survived'
            else:
                self.target = 'Not Survived'
            return
        self.left = Decision_tree(cur_depth = self.depth + 1)
        self.left.train_dt(left)

        self.right = Decision_tree(cur_depth = self.depth + 1)
        self.right.train_dt(right)
        return

    def predict(self, test):
        if test[self.feat] > self.value:
            if self.right is None:
                return self.target
            return self.right.predict(test)        
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test)
        
```

```python
d_t = Decision_tree()
```
